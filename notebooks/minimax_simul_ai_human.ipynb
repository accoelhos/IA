```python
from tic_tac_toe import TicTacToe
from minimax import minimax, best_move

def play_vs_ai():
    game = TicTacToe()
    human = input("Você quer ser X ou O? ").upper().strip()
    assert human in ['X', 'O']
    ai = 'O' if human == 'X' else 'X'

    while not game.game_over():
        print("\nTabuleiro:")
        print(game.board[0:3])
        print(game.board[3:6])
        print(game.board[6:9])

        if game.current == human:
            while True:
                try:
                    move = int(input(f"Sua vez ({human}). Escolha posição (0-8): "))
                    if move in game.available_moves():
                        game.make_move(move)
                        break
                    else:
                        print("Posição inválida.")
                except ValueError:
                    print("Digite um número entre 0 e 8.")
        else:
            move = best_move(game)
            print(f"IA ({ai}) escolheu a posição {move}.")
            game.make_move(move)

    print("\nTabuleiro final:")
    print(game.board[0:3])
    print(game.board[3:6])
    print(game.board[6:9])

    result = game.winner()
    if result == human:
        print("Você venceu!")
    elif result == ai:
        print("A IA venceu.")
    else:
        print("Empate.")

play_vs_ai()
```
